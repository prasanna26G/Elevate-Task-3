```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('ggplot')
```

```python
netflix_df = pd.read_csv('/content/netflix_titles.csv')
netflix_df.head()
```

```python
from sklearn.datasets import load_iris
import pandas as pd

iris = load_iris(as_frame=True)
iris_df = iris.frame
```

```python
netflix_df.info()
netflix_df.describe()
netflix_df.isnull().sum()
```

```python
iris_df.info()
iris_df.describe()
```

```python
iris_df.hist(figsize=(10,6))
plt.suptitle("Distribution of Numerical Features - Iris Dataset")
plt.show()
```

```python
sns.countplot(data=netflix_df, x='type')
plt.title("Movies vs TV Shows on Netflix")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(data=netflix_df, y='rating', order=netflix_df['rating'].value_counts().index)
plt.title("Content Ratings Distribution")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(data=iris_df.drop(columns='target'))
plt.title("Outlier Detection - Iris Dataset")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(iris_df.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap - Iris Dataset")
plt.show()
```

```python
iris_df.corr()['target'].sort_values(ascending=False)
```

###  Insight:
- Petal features are highly correlated
- Sepal width shows weak correlation with target

**Why detect outliers?**

**What is multicollinearity?**
